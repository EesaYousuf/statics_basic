```python
# Different Tests to check Normaity of the Data 1 : shapiro 2 : Normal Test 3 : Anderson
```

```python
from scipy import stats
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/sunnysavita10/Statistics-With-Python-CompleteGuide/refs/heads/main/Iris.csv")
stats.shapiro(df['SepalWidthCm'])
```

```python
# Null_hypothesis is that my data is normally Distribured
p_value=0.07517918532015967
```

```python
if p_value<=0.05:
    print("Reject Null Hypotheiss")
else:
    print("Accept null hypotheisis")
```

```python
import numpy as np
from scipy import stats
d=np.random.lognormal(3,1,1000)
stats.normaltest(np.log(d))
```
