```python
import numpy as np
import seaborn as sns
# Normal Distribution

s=np.random.normal(0.5,0.2,1000)
sns.histplot(s,kde=True)
```

```python
# Log Normal Distribution (Right Skewed Data)
d=np.random.lognormal(3,1,1000)    # first param goes mean second as standard Deviation(sigma)
sns.displot(d,kde=True)
```

```python
# Now if we apply log on top of the lognormal data it will form Normal Distribution now lets see this visually and prove it

sns.histplot(np.log(d),kde=True)
```
